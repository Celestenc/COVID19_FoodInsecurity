##Cleaning 2019 data
Resulting dataframe **rank2019**



### 1st sheet: Outcomes & Factors Rankings

```python
df1=pd.read_excel("2019_health_r.xlsx",sheet_name='Outcomes & Factors Rankings')
df1
```

```python
#Deleting colums
df1c=df1.drop(columns=df1.iloc[:,7:19])
#Deleting rows
df1c=df1c.drop([62,63])

#Renaming columns
new_col1={'Unnamed: 0': 'fips',
         'Unnamed: 1': 'state',
         'Unnamed: 2': 'county_name',
         'Health Outcomes': 'health_outcomes_z',
         'Unnamed: 4': 'health_outcomes_r',
         'Health Factors': 'health_factors_z',
         'Unnamed: 6': 'health_factors_r'}
df1c=df1c.rename(columns=new_col1)
df1c
```

### 2nd Sheet: Outcomes & Factors SubRankings

```python
df2=pd.read_excel("2019_health_r.xlsx",sheet_name='Outcomes & Factors SubRankings')
df2
```

```python
#Remove columns
to_remove=df2.iloc[:,np.r_[1:3,7:15]]
df2c=df2.drop(columns=to_remove)

#Remove rows
df2c=df2c.drop([0,1])

#Rename columns
new_col2={'Unnamed: 0':'fips',
          'Length of Life':'lenght_life_z',
          'Unnamed: 4':'length_life_r',
          'Quality of Life':'quality_life_z',
          'Unnamed: 6':'quality_life_r'}
df2c=df2c.rename(columns=new_col2)
df2c
```

### 3rd Sheet: Ranked Measure Data

```python
df3=pd.read_excel("2019_health_r.xlsx",sheet_name='Ranked Measure Data')
df3
```

```python
#Deleting columns
to_remove=df3.iloc[:,np.r_[1:3,4:10,11:34,35:38,39:98,100:103,104:162]]
df3c=df3.drop(columns=to_remove)

#Deleting rows
df3c=df3c.drop([0,1])

#Renaming columns
new_col3= {'Unnamed: 0':'fips',
          'Premature death':'yppl',
          'Poor or fair health':'%poor_fair_health',
          'Adult obesity':'%adult_obesity',
          'Food environment index':'food_environment_index',
          'High school graduation': 'graduation_cohort',
          'Unnamed: 99': 'graduation_rate',
          'Unnamed: 103':'%some_college'}

df3c=df3c.rename(columns=new_col3)
df3c
```

### 4th Sheet: Additional Measure Data




```python
df4=pd.read_excel("2019_health_r.xlsx",sheet_name='Additional Measure Data')
df4
```

```python
#Remove columns
to_remove=df4.iloc[:,np.r_[1:3,4:36,37:42,43,45:65,66:71,72:93,94,96,98,100,102,104:111]]
df4c=df4.drop(columns=to_remove)

#Remove rows
df4c=df4c.drop([0,1])

#Rename columns
new_col4={'Unnamed: 0': 'fips',
          'Life expectancy':'life_expectancy',
          'Diabetes prevalence':'%diabetes',
          'Unnamed: 42':'%food_insecure',
          'Unnamed: 44':'%limited_access',
          'Median household income':'household_income',
          'Children eligible for free or reduced price lunch':'%lunch_aid',
          'Unnamed: 93':'%african_american',
          'Unnamed: 95':'%american_indian',
          'Unnamed: 97':'%asian',
          'Unnamed: 99':'%pacific_islander',
          'Unnamed: 101':'%hispanic',
          'Unnamed: 103':'%white'
    
}
df4c=df4c.rename(columns=new_col4)
df4c
```

### Merge 4 sheets


```python
#Merge first 2
merged_df1=pd.merge(df1c,df2c,on='fips',how='outer')

#Merge second 2
merged_df2=pd.merge(df3c,df4c,on='fips',how='outer')

#Merge all together
rank2019=pd.merge(merged_df1,merged_df2,on='fips',how='outer')

#Keep only counties in NYC
#county_nyc=['New York','Kings','Bronx','Richmond','Queens']
#rank2019=rank2019[rank2019['county_name'].isin(county_nyc)]
rank2019=rank2019.reset_index().drop(columns=['index','health_outcomes_z','health_factors_z','lenght_life_z','quality_life_z'])
rank2019
```

```python
#change data types
floats=['yppl','%poor_fair_health','%adult_obesity','food_environment_index','graduation_rate','%some_college','life_expectancy','%diabetes','%food_insecure',
        '%limited_access','%lunch_aid','%african_american','%american_indian','%asian','%pacific_islander','%hispanic','%white']

ints=['fips','health_outcomes_r','health_factors_r','length_life_r','quality_life_r','graduation_cohort','household_income']

rank2019[floats]=rank2019[floats].astype(float)
rank2019[ints]=rank2019[ints].astype(int)

#Save and export
pd.set_option("display.precision", 4)
rank2019
rank2019.to_csv('rank2019.csv',encoding='utf-8',index=False)
```
